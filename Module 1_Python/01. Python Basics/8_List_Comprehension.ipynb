```python
list1=[64, 125, 216, 343, 512, 729]
list2=[16,25,36,49,64,81]

subtracted=[]
for item1, item2 in zip(list1, list2):
  subtracted.append(item1-item2)
print(subtracted)

#output: [48, 100, 180, 294, 448, 648]
```
